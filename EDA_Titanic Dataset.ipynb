```python
# function to determine the missing fields and its percentage
def findmissing(df1):
     missing_item=df1.isnull().sum().sort_values(ascending=False)[df.isnull().sum().sort_values(ascending=False)!=0]
     missing_percentage=round((df1.isnull().sum()/df.isnull().count())*100,2)[round((df.isnull().sum()/df.isnull().count())*100,2)!=0]
     missingdata=pd.concat([missing_item,missing_percentage],axis=1,keys=['Missing Field','Missing Percentage'])
     return missingdata 
```

```python
# calling the function to display the missing item.
findmissing(df).style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
# function to handle missing values
def fillage(cols):
    Age=cols[0]
    Pclass=cols[1]
    if(pd.isnull(Age)):
        if (Pclass)==1:
            return 38
        elif (Pclass)==2:
            return 29
        else:
            return 24 
    else:
        return Age
        
            
    
```

```python
# calling the function 
df["Age"]=df[["Age","Pclass"]].apply(fillage,axis=1)
```
